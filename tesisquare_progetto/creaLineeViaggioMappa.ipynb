```python
from geopy.geocoders import Nominatim

def get_coordinates_from_cap(cap, country):
    """
    Ottiene le coordinate (lat, lon) da un CAP.
    
    :param cap: Codice postale (stringa o numero)
    :param country: Codice del paese (default: Italia -> "IT")
    :return: Tuple (latitudine, longitudine) o None se non trovato
    """
    geolocator = Nominatim(user_agent="cap_to_coordinates")
    location = geolocator.geocode({"postalcode": cap, "country": country})
    
    if location:
        return (location.latitude, location.longitude)
    else:
        return None
        
viaggi=[]#inserisco le cordinate della partenza, coordinarte dell'arrivo

for _, row in dataSet.iterrows():
    partenza = get_coordinates_from_cap(row["DEPARTURE_ZIPCODE"], row["DEPARTURE_COUNTRY"])
    arrivo = get_coordinates_from_cap(row["ARRIVAL_ZIPCODE"], row["ARRIVAL_COUNTRY"])

    
    '''
# Esempio di utilizzo
cap = "62010"  # CAP di Roma
coordinate = get_coordinates_from_cap(cap)

if coordinate:
    print(f"Coordinate di {cap}: {coordinate}")
else:
    print(f"CAP {cap} non trovato.")
'''
print(viaggi)
```
