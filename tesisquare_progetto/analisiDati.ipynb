```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("./validated_delivery_data_ml_imputed.csv")

dataset.head()
```

```python
DataSetLocal = dataset.query("DEPARTURE_COUNTRY == ARRIVAL_COUNTRY")
DataSetInter = dataset.query("DEPARTURE_COUNTRY != ARRIVAL_COUNTRY")

DataSetLocal.to_csv("datiStessoPaese.csv",index=False)
DataSetInter.to_csv("datiPaeseDiverso.csv",index=False)
```

```python
fig, ax=plt.subplots(figsize=(6, 6))
#ax.figure()
sns.boxplot(x=dataset["DELIVERY_TIME_HH"])
ax.set_title("boxplot tempo viaggio[h]")
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(6, 6))
#ax.figure()
sns.boxplot(x=dataset["DECLARED_DISTANCE_KM"])
ax.set_title("boxplot km viaggio")
plt.show()
```

```python
fig2, ax2=plt.subplots(figsize=(7,7))
ax2.scatter(dataset["DECLARED_DISTANCE_KM"],dataset["DELIVERY_TIME_HH"],color="red", s=5)
plt.show()
```
